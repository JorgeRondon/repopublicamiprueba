# <center>¿Que es una CPU?</center><br>

La **CPU (Central Processing Unit)** o unidad central de proceso, es uno de los elementos principales que hace parte de la gran mayoria de dispositivos electrónicos. El papel de esta unidad en los dispositivos se suele relacionar con la función que cumple el cerebro en el cuerpo humano, ya que es quien recibe las señales provenientes de diferentes fuentes, las analiza y genera señales de respuesta que luego se traducen en acciones que realiza el sistema controlado por la CPU [1].

![CPU](https://i.blogs.es/ddda0d/procesador/1366_2000.jpg "CPU: Unidad Central de Proceso. Tomado de [1]")

Para el caso especifico de los dispositivos electrónicos, la CPU es quien se encarga de procesar la información proveniente del sistema operativo y tambien de las aplicaciones que esta corriendo la maquina. La velocidad a la cual el sistema ejecuta las tareas depende de varios aspectos relacionados con la construcción de la Unidad Central de Proceso, entre los cuales se encuentran:

> 1. La cantidad de nucleos de la CPU.
> 2. La cantidad de hilos simultaneos que puede procesar cada núcleo.
> 3. La velocidad (medida en Hz) de cada nucleo.


# <center>¿Que es una GPU?</center><br>

La **GPU (Graphic Processing Unit)** o unidad de procesamiento grafico, principalmente usada cuando es necesario el procesamiento de imagenes, como en los video juegos, en la animación por computador, entre otras.

![GPU](https://i.blogs.es/9777ce/chip/1366_2000.jpg "GPU: Unidad de procesamiento grafico. Tomado de [2]")

# <center>¿Que es una TPU?</center><br>

La **TPU (Tensor Processing Unit)** o unidad de procesamiento de tensores, diseñada especificamente para el desarrollo de modelos de aprendizaje profundo, por lo que esta optimizada para realizar operaciones matematicas con matrices de manera simultanea.

![TPU](https://cloud.google.com/images/products/tpu/cloud-tpu-v3-img.png?hl=es-419 "TPU: Unidad de procesamiento de tensores. Tomado de [3]")

# La CPU en la inteligencia artificial

La gran variedad de procesos que puede llevar a cabo una CPU en un dispositivo electrónico la hacen excelente cuando el sistema va a realizar gran variedad de tareas pero no se necesita un desempeño alto especificamente en alguna de ellas. 
Para el caso de la inteligencia artificial, es común que se requiera realizar una gran cantidad de sumas y multiplicaciones duante la ejecución de un proceso, operaciones matematicas que se llevan a cabo en una parte de la CPU que se conoce como **ALU** (Unidad logica aritmetica). El principal problema que se presenta al usar CPU en procesos relacionados con la IA es que ella solo puede procesar una cantidad pequeña de estas operaciones simultaneamente, ya que posee pocas ALUs en su construcción, lo que hace que los tiempos de ejecución sean muy altos.

# La GPU en la inteligencia artificial

La GPU es mas poderosa que la CPU, ya que su arquitectura esta diseñada para procesar mas eficientemente datos complejos como graficos en 2 y 3 dimensiones y secuencias de video.
 
La GPU posee una cantidad de ALUs mayor que una CPU, lo que le permite ser hasta miles de veces mas rapidas que la CPU y por ende puede realizar una gran cantidad de operaciones matematicas (sumas y multiplicaciones) simultaneamente. Es por esta razon que la GPU es mas adecuada que una CPU para el desarrollo de Inteligencia Artificial. Sin embargo, la GPU sigue siendo de proposito general comparada con otras unidades de procesamiento como las TPU.

# La TPU en la inteligencia artificial

La TPU ha sido diseñada para trabajar en inteligencia artificial, esto hace que sea la mejor opción para procesar inteligencia artificial. Sin embargo, estas unidades de procesamiento de tensores han sido diseñadas por google y no estan a la venta, por lo que su uso se restringe al trabajo en google colaboratory, en donde google le presta sus TPUs para que usted corra sus modelos.
